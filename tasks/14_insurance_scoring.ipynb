## Взаимная корреляция биометрии

```python
data = pd.DataFrame(
    train[train["Response"] == 1],
    columns=["Wt", "Ht", "Ins_Age", "BMI"]
)
sns.pairplot(data, height=4)
plt.show()
del data
```

## Кластеризация по биометрии

```python
columns_groups = [
    ["Wt", "Ht"],
    ["Wt", "Ins_Age"],
    ["Wt", "BMI"],
    ["Ht", "Ins_Age"],
    ["Ht", "BMI"],
    ["Ins_Age", "BMI"],
]

colors = [
    "#A62639",
    "#BD324D",
    "#56494E",
    "#A29C9B",
    "#511C29",
    "#0000FF",
    "#FF00FF",
    "#FFFF00",
    "#00FFFF",
    "#00FF00",
]

fig = plt.figure(figsize=(18, 12))
for i, c in enumerate(columns_groups):
    data = pd.DataFrame(train, columns=c.append("Response"))
    legend = []
    area = fig.add_subplot(2, 3, i+1)
    for response in range(1, train["Response"].max() + 1):
        data[data["Response"] == response].plot.scatter(x=c[0], y=c[1], ax=area, c=colors[response-1])
        legend.append(response)
    area.legend(legend)

plt.show()
```
